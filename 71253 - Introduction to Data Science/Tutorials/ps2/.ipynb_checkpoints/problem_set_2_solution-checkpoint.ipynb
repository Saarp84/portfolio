## 1. Visualizing the OKCupid dataset

### 1.1 Plot the distribution of "drinks" variable

```python
%pylab inline
import pandas as pd
import seaborn as sns
sns.set(style='ticks', font_scale=2, rc={"lines.linewidth": 3})
```

```python

## read the data from the file "okcupid_profiles_ps1.csv" into a pandas dataframe.
dfile = 'okcupid_profiles_ps1.csv'
data = pd.read_csv(dfile)
```

```python
d_drinks = data.drinks
d_drinks.value_counts().plot(kind='bar')
```

> The plot below shows the counts, not the proportions of the different categories. 
We could normalize ourselves, but it's easier to use the 'seaborn' package, as shown below.

### 1.2 Plot the distribution of "drinks" variable stratified by sex (on a single graph) 

```python
ax = sns.histplot(data=data, x='drinks', hue='sex', 
             multiple="dodge", stat = 'count', discrete=True, common_norm=False, shrink=0.5 )
setp(ax.get_xticklabels(), "rotation", 90);
```

```python
ax = sns.histplot(data=data, x='drinks', hue='sex', 
             multiple="dodge", stat = 'probability', discrete=True, common_norm=False, shrink=0.5 )
setp(ax.get_xticklabels(), "rotation", 90);
```

> Above, I've normalized the male and female categories separately (such that each sums to 1) using 'common_norm=False'. By default, the 'histplot' function normalizes all the data together (such that the fractions of male + female sum to 1). Since there are many more males than female in this dataset, a common normalization makes it harder to compare which 'drinks' category is more prevalent between males and female. 

### 1.3 Based on your plot, do you think men and women differ in the drinking habits?

There are more men who drink socially. However, there are more men in the dataset in general. 
Normalizing the data shows that the probability that a woman drinks socially is actually higher than that of a man.   
In geneal, the distribution seems very similar between men and women.  

As always, we need to be cognecent of the fact that these data represnt a very specific subpopulation, and are self reported. So generalizing based on these data is hard.

## 2. Typical values

### 2.1 What are the mean, median, and mode of the 'drinks' variable of the OKCupid users?

Treat 'drinks' as an ordinal variable, with the following order: ['desperately', 'very often', 'often', 'socially', 'rarely', 'not at all']  
> Note that you'll have to encode this variable using numeric values

```python
d_drinks = data.drinks.dropna()
drinks_mode = d_drinks.value_counts().index[0]

drinks_num = pd.Series(range(6)[::-1], index=['desperately', 'very often', 'often', 'socially', 'rarely', 'not at all'])

d_drinks_num = d_drinks.apply(lambda x:drinks_num[x])

drinks_median = drinks_num[drinks_num==d_drinks_num.median()].index[0]

print('The mean "drinks" is not defined since this is an ordinal variable' )
print('The median  "drinks" is {:}'.format(drinks_median))
print('The mode  "drinks" is {}'.format(drinks_mode))
```

### 2.2 What are the mean, median, and mode of the income values of the OKCupid users?

```python
d_income = data.income[data.income>=0] #restricting to non-negative values only, since there can't be negative income
income_stats = d_income.describe()
mode = d_income.value_counts().index[0]
print('The mean income {:g}'.format(income_stats['mean']))
print('The median income {:g}'.format(income_stats['50%']))
print('The mode income {:g}'.format(mode))
```

```python
d_income.hist()
```

Note that there are several high income values of 1e6, with a gap from the nearest closest values.  
These values are suspect to be a generic high number, rather than an exact value.  
Nonetheless, I chose to retain them.

#### What do these statistics tell you about the shape of the income distriubiton in general population?

Since the OKCupid users are not a random sample of the population, it's risky to generalize from these data to the general populaiton.  

#### Plot the distribution of income variable. Indicate the mean mode and median the plot.

***

```python
ax = sns.histplot(data=d_income, log_scale=True, kde=True, stat = 'probability')

ylm = [0, 0.7]

ax.plot([income_stats['mean']]*2, ylm, '--k', label='mean')
ax.plot([income_stats['50%']]*2, ylm, '-.', c='grey', label='median')
ax.plot([mode]*2, ylm, ls='dotted', color='darkolivegreen', label='mode')

ax.set_ylim(ylm)

legend()
```

> Above, the meaning of the different lines is given in the caption, which is off to the side. I prefer it when the annotations are right next to the elements they are annotating. One option for how to do so is shown below.

```python
ax = sns.histplot(data=d_income, log_scale=True, kde=True, stat = 'probability')

ylm = [0, 0.7]

ax.plot([income_stats['mean']]*2, ylm, '--k', label='mean')
ax.plot([income_stats['50%']]*2, ylm, '-.', c='grey', label='median')
ax.plot([mode]*2, ylm, ls='dotted', color='darkolivegreen', label='mode')

fs=16
ax.text(mode, ylm[1]+0.01, 'mode', va='bottom', ha='center', fontsize=fs, weight='bold', color='darkolivegreen') 
ax.text(income_stats['50%'], ylm[1]+0.01, 'median', va='bottom', ha='center', fontsize=fs, weight='bold', color='grey') 
ax.text(income_stats['mean'], ylm[1]+0.01, 'mean', va='bottom', ha='center', weight='bold', fontsize=fs, color='k') 

ax.set_ylim(ylm)
```
