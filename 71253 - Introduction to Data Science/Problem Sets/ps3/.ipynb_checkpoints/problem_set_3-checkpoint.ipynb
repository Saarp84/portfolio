## 2. The typical man

We're now interested in what does a 'typical' man look like. For that we'll look for men who are within a few standard deviations of the sample mean in their measurements.  

### 2.1 Observed fraction of typical men

First, we'll consider individual variables (=body measurements). We'll focus on the variables: 'Heightin', 'Weightlbs', 'headcircumference', 'handlength'.  
For each variable, we're interested in what fraction of men are "typical" in that variable. (E.g., what fraction of men are of "typical" height?) 

In particular, compute the fraction of the men in dataset are within 1, 2, or 3 standard deviations of the sample mean of each variable.  
(For example, what fraction of men are within 1 standard deviation in height from the sample mean height?) 

```python
def sampleprop(df,v,g=['Male','Female']):
    
    df1=df[df.Gender.isin(g)]
    mean = df1[v].mean()
    std = df1[v].std()
    print(f"As for the variable {v} and gender group {g}")
    range1 = [mean-std,mean+std]
    d1 = df1[(df1[v]>=range1[0]) & (df1[v]<=range1[1])]
    print(f"{len(d1)/len(df1):.1%} of the sample lies within 1 standard deviation.")
    range2 = [mean-2*std,mean+2*std]
    d2 = df1[(df1[v]>=range2[0]) & (df1[v]<=range2[1])]
    print(f"{len(d2)/len(df1):.1%} of the sample lies within 2 standard deviations.")
    range3 = [mean-3*std,mean+3*std]
    d3 = df1[(df1[v]>=range3[0]) & (df1[v]<=range3[1])]
    print(f"{len(d3)/len(df1):.1%} of the sample lies within 3 standard deviations.")
    
#     cv= std/mean
#     print(f"Coefficient of varaiation is: {cv:.2}")
```

```python
sampleprop(df,'Heightin',['Male'])
```

```python
sampleprop(df,'Weightlbs',['Male'])
```

```python
sampleprop(df,'headcircumference',['Male'])
```

```python
sampleprop(df,'handlength',['Male'])
```

#### 2.3 Fraction of "completely" typical men

Next, we're interested in how many men are typicall in all their primary body measuremnts (=completely typical).  

2.3.1 First, compute the fraction of the men in dataset that are within 1 standard deviations of the sample mean of all the variables we've considered in the previous section.  
(I.e. what fraction of men are typical in their height, weight, head circumference, and hand length?)

```python
var_list = ['Heightin', 'Weightlbs', 'headcircumference', 'handlength']
```

```python
def std1(df,v1):
    m=df[v1].mean()
    s=df[v1].std()
    r=[m-s,m+s]
    return r

df1 = df[df['Gender']=='Male']
d_filttered = df1[(df1['Heightin']>=std1(df1,'Heightin')[0]) &
                  (df1['Heightin']<=std1(df1,'Heightin')[1]) &
                  (df1['Weightlbs']>=std1(df1,'Weightlbs')[0]) &
                  (df1['Weightlbs']<=std1(df1,'Weightlbs')[1]) &
                  (df1['headcircumference']>=std1(df1,'headcircumference')[0]) & 
                  (df1['headcircumference']<=std1(df1,'headcircumference')[1]) &
                  (df1['handlength']>=std1(df1,'handlength')[0]) &
                  (df1['handlength']<=std1(df1,'handlength')[1])]

p=len(d_filttered)/len(df1)

print(f"{p:.2%} of the sample lies within 1 standard deviation (completely typical men)")
```

2.3.2 Calculate the **expected** fraction of men that are typical in all of these 4 variables. 
(I.e. ones that are within 1 standard deviation in all of these variables.)
  
> Assume that each variable is normaly distributed and that all variables are independent.

```python
# for one noramal distributed variable fraction is 68.26%
# for 4 variables together:
print(f"{0.6826**4:.2%}")
```

2.3.4 But there are many more body measurements per person. Now compute the same fraction, but considering all the major body size measurements (given by the first 91 variables).  
(I.e. what fraction of men are typical in all 91 of their body measurements?)

```python
df1 = df[df['Gender']=='Male']
vlist = df.columns.tolist()[1:92]

print(f"Initial number of rows for men: {len(df1)}")

l = df1[(df1[vlist[0]] >= std1(df, vlist[0])[0]) &
        (df1[vlist[0]] <= std1(df, vlist[0])[1])]['InnerIndex'].unique().tolist()
print(f"Initial filter for {vlist[0]}: {len(l)} unique indices")

for i in range(1, len(vlist)):
    indices = df1[(df1[vlist[i]] >= std1(df, vlist[i])[0]) &
                  (df1[vlist[i]] <= std1(df, vlist[i])[1])]['InnerIndex'].unique().tolist()

    print(f"Filter for {vlist[i]}: {len(indices)} unique indices before intersection")

    l = list(set(l) & set(indices))

    print(f"After intersection number {i} with {vlist[i]}: {len(l)} unique indices")
    print("-----------------------------------------------------------------------")
```

```python
df2 = df1[df1['InnerIndex'].isin(l)]
p = len(df2) / len(df1)

print(f"Final filtered data: {len(df2)} rows")
print(f"Proportion of sample within 1 standard deviation: {p:.2%}")
```

2.3.5 Why are there so few completely typical men, despite the fact that there are plenty of typical men for each individual variable?

```python
# There are many variables in the data that do not depend on each other, so the probability for the intersections is very small

# with 91 independent normal distributed variables
print(f"with 92 independent normal distributed variables the probabiltiy is {0.6826**92:.16%}")
```
