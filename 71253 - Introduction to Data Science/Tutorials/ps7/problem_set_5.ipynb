## 2. Regressing (continous) variables

### 2.1 Perform a linear regression of BMI against Age (i.e. BMI = intercept + slope*Age) 
### What are the regression's slope and intercept?

```python
from sklearn import linear_model

my_model = linear_model.LinearRegression()
results = my_model.fit(df_filtered[['Age']], df_filtered[['BMI']])
print("The linear model is: BMI = {:.5} + {:.5}*AGE".format(results.intercept_[0],results.coef_[0][0]))
moe = results.coef_[0][0]
```

### 2.2 Generate a scatter plot of the data and the regression line

```python
predictions = results.predict(df_filtered[['Age']])

fig, ax = plt.subplots()
# we can add the predicted values to the original plot
ax = sns.scatterplot(x=df_filtered['Age'],y=df_filtered['BMI'])

ax.plot(df_filtered['Age'],
        predictions,
        color="b")
```

### 2.3 Now we want to use a permutation test to test the null hypothesis that the slope is 0. 
### To generate data corresponding to the null hypothesis, we can permute one variable with respect to the other. 
### Implement such a test using the slope as the test statistic (magnitude of effect). 
### What is the resulting p-value?

```python
data = df_filtered.copy()
runs = 10000
diffs = np.zeros(runs)
for ii in range(runs):
    data['Age_shuffled'] = data['Age'].sample(frac=1, replace=False).values
    my_model = linear_model.LinearRegression()
    results = my_model.fit(data[['Age_shuffled']], data[['BMI']])
    diffs[ii]= results.coef_[0][0]
```

```python
# moe = 0.013103
larger = np.where(diffs>=moe, 1, 0)
p_val = np.sum(larger)/runs
print(f'The p-value is {p_val}')
```
