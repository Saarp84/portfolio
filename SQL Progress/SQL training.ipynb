```python

# Summing up game score for each team
query9_step1 = """
SELECT 
    Tm,
    Data,
    SUM(PTS) AS Tm_score
FROM nba_players
GROUP BY Tm, Data
"""
mydata = pd.read_sql_query(query9_step1, engine)
mydata.head(2)

# Calculating the Average Score per Team
query9_step2 = """
SELECT 
    Tm, 
    AVG(Tm_score) AS avg_score
FROM (
    SELECT 
        Tm,
        Data,
        SUM(PTS) AS Tm_score
    FROM nba_players
    GROUP BY Tm, Data
) AS subquery
GROUP BY Tm
ORDER BY avg_score DESC;
"""
# Group BY  - using for drop duplicates

mydata = pd.read_sql_query(query9_step2, engine)
mydata.head(2)
```

```python
# Step 1: Execute the first query and store the result in a DataFrame
query9_step1 = """
SELECT 
    Tm,
    Data,
    SUM(PTS) AS Tm_score
FROM nba_players
GROUP BY Tm, Data
"""
tm_data = pd.read_sql_query(query9_step1, engine)

# Step 2: Calculate the average score per team using Pandas
avg_score_per_team = tm_data.groupby('Tm')['Tm_score'].mean().reset_index()
avg_score_per_team.columns = ['Tm', 'avg_score']

# Sort the results by avg_score descending
avg_score_per_team = avg_score_per_team.sort_values(by='avg_score', ascending=False)

# Display the result
avg_score_per_team.head()
```

```python
# INNER JOIN
query21 = """
SELECT 
    p.Player,
    p.Tm,
    i.country,
    i.draft_year,
    i.draft_round,
    i.draft_number,
    ROUND(AVG(CASE WHEN p.Data > '2025-01-01' THEN p.PTS END),2) AS avg_pts,
    ROUND(AVG(CASE WHEN p.Data > '2025-01-01' THEN p.AST END),2) AS avg_ast,
    ROUND(AVG(CASE WHEN p.Data > '2025-01-01' THEN p.TRB END),2) AS avg_reb,
    COUNT(DISTINCT CASE WHEN p.Data > '2025-01-01' THEN p.Data END) AS games_played
FROM nba_players p
INNER JOIN players_info i
    ON p.Player = i.player_name
WHERE p.Tm = 'POR'
AND p.Data > '01/01/2025'
GROUP BY p.Player, p.Tm, i.country, i.draft_year, i.draft_round, i.draft_number
ORDER BY avg_pts DESC

"""

validation_query = """
SELECT p.Player, p.Tm, p.PTS, p.AST, p.TRB, p.Data
FROM nba_players p
WHERE p.Player = 'Bryce McGowens'
AND p.Data > '2025-01-01'
ORDER BY p.Data
"""




mydata = pd.read_sql_query(query21, engine)
mydata
```

```python
query22 = """
SELECT
    p.Player,
    COUNT(DISTINCT CASE WHEN p.Res = 'W' THEN p.Data END) AS W,
    COUNT(DISTINCT CASE WHEN p.Res = 'L' THEN p.Data END) AS L,
    ROUND(AVG(CASE WHEN p.Res = 'W' THEN p.PTS END), 2) AS avg_pts_W,
    ROUND(AVG(CASE WHEN p.Res = 'L' THEN p.PTS END), 2) AS avg_pts_L,
    ROUND(AVG(CASE WHEN p.Res = 'W' THEN p.MP END), 2) AS avg_minplayed_W,
    ROUND(AVG(CASE WHEN p.Res = 'L' THEN p.MP END), 2) AS avg_minplayed_L
FROM nba_players p
WHERE p.Tm = 'POR'
GROUP BY p.Player
HAVING avg_pts_W > 5
ORDER BY avg_minplayed_W DESC
"""

mydata = pd.read_sql_query(query22, engine)
mydata
```

```python
ex_query1 = """
SELECT
    p.Player,
    COUNT(DISTINCT CASE WHEN p.Data > '2024-12-31' THEN p.Data END) AS games_played,
    COUNT(DISTINCT CASE WHEN p.PTS >= 20 and p.Data > '2024-12-31' THEN p.Data END) AS games_above20,
    COUNT(DISTINCT CASE WHEN p.PTS < 20 and p.Data > '2024-12-31' THEN p.Data END) AS games_below20
FROM nba_players p
WHERE p.Tm = 'POR'
GROUP BY p.Player
HAVING games_played > 5
ORDER BY games_above20 DESC
"""
fix_query1 = """
SELECT 
    p.Player,
    COUNT(DISTINCT p.Data) AS games_played,
    COUNT(DISTINCT CASE WHEN p.PTS >= 20 THEN p.Data END) AS games_above20,
    COUNT(DISTINCT CASE WHEN p.PTS < 20 THEN p.Data END) AS games_below20
FROM nba_players p
WHERE p.Tm = 'POR' 
AND p.Data > '2024-12-31' 
GROUP BY p.Player
HAVING games_played > 5
ORDER BY games_above20 DESC;
"""
mydata = pd.read_sql_query(fix_query1, engine)
mydata
```

```python

ex_query2 = """
SELECT
    p.Player,
    ROUND(AVG(CASE WHEN p.Res = 'W' THEN p.PTS END),2) AS avg_pts_W,
    ROUND(AVG(CASE WHEN p.Res = 'L' THEN p.PTS END),2) AS avg_pts_L,
    ROUND(AVG(CASE WHEN p.Res = 'W' THEN p.PTS END) - AVG(CASE WHEN p.Res = 'L' THEN p.PTS END), 2) AS pts_diff
FROM nba_players p
GROUP BY p.Player
HAVING avg_pts_W > 15
ORDER BY pts_diff desc
"""

mydata = pd.read_sql_query(ex_query2, engine)
mydata
```

```python
ex_query3 = """
select 
    p.Player,
    round(avg(p.PTS),2) as avg_pts,
    RANK() OVER (ORDER BY avg(p.PTS) DESC) as pts_rank
from nba_players p
group by p.Player
order by pts_rank
"""
outer_query3 = """
SELECT 
    sub.Player,
    sub.avg_pts,
    sub.max_pts,
    sub.pts_rank


FROM (
    SELECT 
        p.Player,
        ROUND(AVG(p.PTS), 2) AS avg_pts,
        MAX(p.PTS) AS max_pts,
        RANK() OVER (ORDER BY AVG(p.PTS) DESC) AS pts_rank
    FROM nba_players p
    GROUP BY p.Player
) AS sub

WHERE sub.Player IN (SELECT DISTINCT Player FROM nba_players WHERE Tm = 'POR')
ORDER BY sub.pts_rank;

"""


mydata = pd.read_sql_query(outer_query3, engine)   
mydata
```

```python
query3 = """
select
    film.title,
    category.name,
    count(distinct actor.actor_id) as num_of_actress
from film
JOIN film_category ON film.film_id = film_category.film_id
JOIN category ON film_category.category_id = category.category_id
JOIN film_actor ON film.film_id = film_actor.film_id
JOIN actor ON film_actor.actor_id = actor.actor_id
where category.name = 'Comedy'
GROUP BY film.title, category.name
ORDER BY num_of_actress DESC
"""
query4 = """
select
    category.name,
    count(distinct film.film_id) as num_of_movies
from film
JOIN film_category ON film.film_id = film_category.film_id
JOIN category ON film_category.category_id = category.category_id
JOIN film_actor ON film.film_id = film_actor.film_id
JOIN actor ON film_actor.actor_id = actor.actor_id
GROUP BY category.name
ORDER BY num_of_movies DESC
"""

query5 = """
select
    actor.first_name,  
    actor.last_name,
    count(distinct film.film_id) as num_of_movies
from film
JOIN film_category ON film.film_id = film_category.film_id
JOIN category ON film_category.category_id = category.category_id
JOIN film_actor ON film.film_id = film_actor.film_id
JOIN actor ON film_actor.actor_id = actor.actor_id
GROUP BY actor.first_name, actor.last_name
ORDER BY num_of_movies DESC
LIMIT 5
"""

pd.read_sql_query(query5, engine2)
```
