## 2. Typical values

### 2.1 What are the mean, median, and mode of the 'drinks' variable of the OKCupid users?

Treat 'drinks' as an ordinal variable, with the following order: ['desperately', 'very often', 'often', 'socially', 'rarely', 'not at all']  
> Note that you'll have to encode this variable using numeric values

```python
# array(['socially', 'often', 'not at all', 'rarely', nan, 'very often', 'desperately'], dtype=object)

cleanup_drink_habbits = {"not at all": 1, 
                         "rarely": 2,
                         "socially": 3,
                         "often": 4,
                         "very often": 5,
                         "desperately": 6}

okupid['drinks_encoded'] = okupid['drinks'].map(cleanup_drink_habbits)

```

```python
inverse_d = {v: k for k, v in cleanup_drink_habbits.items()}

drinks_mean = okupid['drinks_encoded'].mean()

drinks_median = okupid['drinks_encoded'].quantile(.5)
median_cat = inverse_d.get(drinks_median)

drinks_mode = okupid['drinks_encoded'].mode()[0]
mode_cat = inverse_d.get(drinks_mode)

print(f"The mean of `drinks_encoded` is {drinks_mean:.3}")
print(f"The median of `drinks_encoded` is {drinks_median}, I can say that the median category is `{median_cat}`")
print(f"The most frequent value of `drinks_encoded` is {drinks_median}, I can say that the most common category is `{mode_cat}`")
```

### 2.2 What are the mean, median, and mode of the income values of the OKCupid users?

#### What do these statistics tell you about the shape of the income distriubiton in general population?

#### Plot the distribution of income variable. Indicate the mean mode and median on the plot.

```python
okupid['income'].describe(percentiles=[0.25,0.5,0.75,0.8,0.85,0.9,.95])
```

```python
print(f"The mean of users income is: {okupid['income'].mean()} $")
print(f"The median of users income is: {okupid['income'].quantile(.5)} $")
print(f"The most frequent value is: {okupid['income'].mode()[0]} $")

print("I assume the missing values ​​are represented by -1 and most of the income values are missing since income cannot be negative")
```

```python
import scipy.stats as stats
from ipywidgets import interact, interactive, fixed

x = 'income'

@interact(bins=(2,200, 10))
def plot_hist(bins):
    fig, ax = plt.subplots(figsize=(8,6))
    sns.histplot(data=okupid, x=x, bins=bins, stat='density')
    ax.set_xlim(-10000, 110000)#okupid['income'].max())
    ax.axvline(x=okupid['income'].mean(), color='red', linestyle='--', linewidth=0.75,)
#     ax.axvline(x=okupid['income'].quantile(.5), color='green', linestyle='--', linewidth=1.5)
    ax.axvline(x=okupid['income'].mode()[0], color='purple', linestyle='--', linewidth=0.75)
# median and mode are equal
    ax.text(okupid['income'].mean(), 0, 'Mean', color='red', fontsize=8, ha='center', va='top', rotation=0)
    ax.text(okupid['income'].mode()[0], 0, 'Median/Mode', color='purple', fontsize=8, ha='center', va='top', rotation=0)

```
