```python
import pandas as pd
import numpy as np
```

```python
simudata = pd.read_excel("dgima1.9.xlsx")
simudata['index'] = range(1, simudata.shape[0]+1)
simudata = simudata[['index'] + list(simudata.columns[:-1])]

simudata1 = simudata.iloc[:, :16]
```

```python
mydata = simudata.iloc[:, 16:]
mydata['index'] = simudata['index']
mydata = mydata[['index'] + list(mydata.columns[:-1])]

categories = ["גימלאות", "SemelGimla:", "ShemGimla:", "MiTarGimla:", "AdTarGimla:", "GovaGimlaTakin:", "SchumGovaGimla:", "YemeyZakaut:", "YemeyNitzul:",
              "נכויות", "Semelnechut:", "ShemNechut:", "AchuzRefui:", "EKosher:", "MiTarNechut:", "AdTarNechut:",
              "מצב משפחתי", "SemelMazavMish:",
              "עיסוקים", "MisparIsuk:", "SemelIsuk:", "ShemIsuk:", "MaamadIsuk:", "MiTarIsuk:", "AdTarIsuk:", "HachnasaTakin:",
              "הכנסות", "MisparIsukHachnasa:", "ShemHachnasa:", "SchumHachnasa:", "MiTarHachnasa:", "AdTarHachnasa:"]

# Create an empty DataFrame with the category and subcategory names as column names
sub_cat_gim = ["SemelGimla:", "ShemGimla:", "MiTarGimla:", "AdTarGimla:", "GovaGimlaTakin:", "SchumGovaGimla:", "YemeyZakaut:", "YemeyNitzul:"]
sub_cat_nec = ["Semelnechut:", "ShemNechut:", "AchuzRefui:", "EKosher:", "MiTarNechut:", "AdTarNechut:"]
sub_cat_isu = ["MisparIsuk:", "SemelIsuk:", "ShemIsuk:", "MaamadIsuk:", "MiTarIsuk:", "AdTarIsuk:", "HachnasaTakin:"]
sub_cat_hac = ["MisparIsukHachnasa:", "ShemHachnasa:", "SchumHachnasa:", "MiTarHachnasa:", "AdTarHachnasa:"]
sub_cat_maz = ["SemelMazavMish:"]

count_matrix = pd.DataFrame(0, index=range(mydata.shape[0]), columns=categories)
indices = pd.DataFrame(index=range(mydata.shape[0]), columns=categories)
```

```python

# Convert the columns of mydata to string type if needed
mydata = mydata.astype(str)

# Initialize indices and count_matrix with zeros
indices = pd.DataFrame(0, index=range(mydata.shape[0]), columns=categories)
count_matrix = pd.DataFrame(0, index=range(mydata.shape[0]), columns=categories)

# Loop through each category
for j, category in enumerate(categories):
    # Find the column indices that contain the current category in each row
    row_indices = np.char.find(mydata.values, category) != -1

    # Store the row indices where the category is found
    indices[category] = ';'.join(map(str, np.where(row_indices)[0]))

    # Count the occurrences of the category in each row
    count_matrix[category] = row_indices.sum(axis=1)
```

```python
count_gim = 0
count_nech = 0
count_maz = 0
count_isuk = 0
count_hach = 0

for i in range(mydata.shape[0]):
    if count_matrix.iloc[i, 1] > count_gim:
        count_gim = count_matrix.iloc[i, 1]
    if count_matrix.iloc[i, 10] > count_nech:
        count_nech = count_matrix.iloc[i, 10]
    if count_matrix.iloc[i, 17] > count_maz:
        count_maz = count_matrix.iloc[i, 17]
    if count_matrix.iloc[i, 19] > count_isuk:
        count_isuk = count_matrix.iloc[i, 19]
    if count_matrix.iloc[i, 27] > count_hach:
        count_hach = count_matrix.iloc[i, 27]
```

```python
simudata1["גימלאות"] = np.nan
# Create a matrix of NaNs to hold the new columns
new_cols = np.empty((mydata.shape[0], count_gim * len(sub_cat_gim)))
new_cols[:] = np.nan
# Loop over the categories and create the new column names
col_names = []
for i in range(1, count_gim + 1):
    for j in range(len(sub_cat_gim)):
        col_names.append(sub_cat_gim[j] + "_" + str(i))
# Add the new columns to the data frame
simudata1 = pd.concat([simudata1, pd.DataFrame(new_cols, columns=col_names)], axis=1)


simudata1["נכויות"] = np.nan
# Create a matrix of NaNs to hold the new columns
new_cols = np.empty((mydata.shape[0], count_nech * len(sub_cat_nec)))
new_cols[:] = np.nan
# Loop over the categories and create the new column names
col_names = []
for i in range(1, count_nech + 1):
    for j in range(len(sub_cat_nec)):
        col_names.append(sub_cat_nec[j] + "_" + str(i))
# Add the new columns to the data frame
simudata1 = pd.concat([simudata1, pd.DataFrame(new_cols, columns=col_names)], axis=1)


simudata1["מצב משפחתי"] = np.nan
# Create a matrix of NaNs to hold the new columns
new_cols = np.empty((mydata.shape[0], count_maz * len(sub_cat_maz)))
new_cols[:] = np.nan
# Loop over the categories and create the new column names
col_names = []
for i in range(1, count_maz + 1):
    for j in range(len(sub_cat_maz)):
        col_names.append(sub_cat_maz[j] + "_" + str(i))
# Add the new columns to the data frame
simudata1 = pd.concat([simudata1, pd.DataFrame(new_cols, columns=col_names)], axis=1)


simudata1["עיסוקים"] = np.nan
# Create a matrix of NaNs to hold the new columns
new_cols = np.empty((mydata.shape[0], count_isuk * len(sub_cat_isu)))
new_cols[:] = np.nan
# Loop over the categories and create the new column names
col_names = []
for i in range(1, count_isuk + 1):
    for j in range(len(sub_cat_isu)):
        col_names.append(sub_cat_isu[j] + "_" + str(i))
# Add the new columns to the data frame
simudata1 = pd.concat([simudata1, pd.DataFrame(new_cols, columns=col_names)], axis=1)


simudata1["הכנסות"] = np.nan
# Create a matrix of NaNs to hold the new columns
new_cols = np.empty((mydata.shape[0], count_hach * len(sub_cat_hac)))
new_cols[:] = np.nan
# Loop over the categories and create the new column names
col_names = []
for i in range(1, count_hach + 1):
    for j in range(len(sub_cat_hac)):
        col_names.append(sub_cat_hac[j] + "_" + str(i))
# Add the new columns to the data frame
simudata1 = pd.concat([simudata1, pd.DataFrame(new_cols, columns=col_names)], axis=1)

```
